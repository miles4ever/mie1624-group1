```python
#amount spent on the primary credit card last month
#creating a dataframe that contains only gender and cardspent colums
cardspent= df[['gender','cardspent']]
cardspent.head()
```

```python
#creating a dataframe for female and male separately only
female_dataframe= cardspent.loc[cardspent['gender']=='Female']
#female_dataframe
male_dataframe= cardspent.loc[cardspent['gender']=='Male']
#male_dataframe
```

```python
#converting male and female dataframes to numpy array
female_array= female_dataframe['cardspent'].to_numpy()
male_array= male_dataframe['cardspent'].to_numpy()
print(female_array, male_array)
type(female_array)
```

Bootstrapping means resampling 

```python
sample= 50
group= 500
n_female= len(female_array)
n_male= len(male_array)
female_avg= []
male_avg =[]
for i in range(group):
    female_cardspent_avg= np.mean(female_array[np.random.randint(n_female,size= 500)])
    male_cardspent_avg= np.mean(male_array[np.random.randint(n_male,size=500)])
    female_avg.append(female_cardspent_avg)
    male_avg.append(male_cardspent_avg)
#print(female_avg)
#print(n_female)
```
